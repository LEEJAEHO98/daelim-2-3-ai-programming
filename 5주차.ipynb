```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

data = {
    "study_hours": [1, 2, 3, 3, 4, 4, 5, 6, 7, 8],
    "attendance": [50, 55, 60, 80, 60, 75, 75, 80, 90, 95],
    "pass_fail": [0, 0, 0, 1, 0, 1, 1, 1, 1, 1]
}

df = pd.DataFrame(data)

# 입력값과 정답 분리
x = df[["study_hours", "attendance"]]
y = df["pass_fail"]

model = DecisionTreeClassifier(random_state=42)
model.fit(x,y)

print("데이터 준비 완료")
print(f"학생 수: {len(df)}명") # 학생 수: %%명
```

# 모델 정확도 

- 학습에 사용한 데이터로 정확도 확인

```python
accuracy = model.score(x,y)
print("모델 정확도:", accuracy) # 0~1 표시됨 -> 0~100%
print(f"모델 정확도 퍼센트:{accuracy * 100}%")
```

```python
# 모델이 어떤 예측을 했는지 확인
predictions = model.predict(x)

# Columns - 공부시간,출석률,실제,예측,결과
result = pd.DataFrame()

result
```
