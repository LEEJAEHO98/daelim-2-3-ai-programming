```python
# 할인율 계산
# discount: 0.0~1.0 (0%~100%)
def calculate_price(price, discount=0):
    discounted = price * (1- discount)
    return discounted

print(f"기본 가격: {calculate_price(10000)}원")
print(f"할인 가격: {calculate_price(10000, 0.1)}원")
```

# 함수에서 여러값 반환하기

```python
# 최솟값과 최댓값을 동시에 반환하는 함수
def min_max(numbers):
    minimum = min(numbers)
    maximum = max(numbers)
    return minimum, maximum

scores = [10,20,30,40,50,60,70,80]    
low, high = min_max(scores)
print("최솟값:", low)
print("최댓값:", high)
print(min_max(scores))
```

## 문제

class_scores = [45, 62, 78, 55, 90, 83, 71, 66, 48, 95]

1. 성적 분석 함수 생성
2. 아래와 같은 로그 출력 - 4줄
    - 평균: x.x점 (소수점 1번째자리까지만 출력)
    - 최고점: xx점
    - 최저점: xx점
    - 합격(60점 이상): x명

```python
def analyze_scores(scores):
    
    average = sum(scores) / len(scores)
    
    maximum = max(scores)
    minimum = min(scores)
    
    pass_count = sum(1 for score in scores if score >= 60)
    
    
    return average, maximum, minimum, pass_count

class_scores = [45, 62, 78, 55, 90, 83, 71, 66, 48, 95]

avg, high, low, passed = analyze_scores(class_scores)

print(f"- 평균: {avg:.1f}점")
print(f"- 최고점: {high}점")
print(f"- 최저점: {low}점")
print(f"- 합격(60점 이상): {passed}명")
```

# Pandas 심화

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
# 학생 데이터
grade_data = {
    "name": ["kim", "lee", "park", "cho", "choi", "jung"],
    "korean": [88, 72, 95, 65, 80, 91],
    "math": [75, 64, 55, 12, 98, 58],
    "english": [97, 100, 70, 80, 46, 67],
}

grade_df = pd.DataFrame(grade_data)
grade_df
```

# 정렬하기

```python
print(" ---- 국어 점수 낮은 순 ----")
# 화면에 표를 출력하기 위해 display() 함수로 감싸줍니다.
display(grade_df.sort_values("korean")) 

print()

print(" ---- 수학 점수 높은 순 ----")
# 마지막 줄이라도 일관성을 위해 display()로 감싸주는 것이 좋습니다.
display(grade_df.sort_values("math", ascending=False))
```

# 새로운 열 추가하기

```python
grade_df["average"] = \
    (grade_df["korean"] + \
      grade_df["math"] + \
        grade_df["english"]) / 3
grade_df["average"] = grade_df["average"].round(1)
grade_df
```

```python
grade_df["grade"] = grade_df["average"].apply(
    lambda x: "A" if x >= 80 else "B"
)
grade_df
```

# 그룹별 통계

```python
group_avg = grade_df.groupby("grade")["average"].mean()
print("등급별 평균:")
print(group_avg)
```
