복습 문제 : 함수

```python
def calculate_average(numbers):
    total = 0
    for n in numbers:
        total +=n
    return total / len(numbers)
    
my_scores = [80, 90, 70, 100]
avg= calculate_average(my_scores)
print("평균", avg)

```

복습 문제 -종합

Q)3명의 학생 이름과 점수가 있고 점수에 따라 등급을 매기는 함수를 생성하고 로그를 출력합니다.

EX)로그 출력

A학생은 a등급입니다.
B학생은 c등급입니다.
C학생은 b등급입니다.

```python

fstudents = ["A", "B", "C"]
students_scores = [95, 73, 58]

def print_student_grades():
    for name, score in zip(students, students_scores):
        if score == 95:
            grade = "a"
        elif score == 73:
            grade = "c"
        elif score == 58:
            grade = "b"
        else:
            grade = "f"
        
        print(f"{name}학생은 {grade}등급입니다.")

# 함수 실행
print_student_grades()
```

# 머신러닝

>정의: 사람이 규칙을 직접 생성하는 대신,

데이터를 보고 컴퓨터가 스스로 규칙을 찾아내는 방법

일반 프로그래밍 vs 머신러닝

| 구분             | 일반 프로그래밍 | 머신러닝                   |
| ---              | ---            | --                        |
|규칙을 누가 만드냐  | 사람이 직접 만든다 | 컴퓨터가 데이터에 찾는다 |
|필요한 것          |  규칙(if문 등)     | 데이터 + 정답          |
|예시              | if 점수 >= 60:합격  | 과거 데이터를 참조, 기준을 찾게함    |
|장점              | 규칙이 적거나,명확할때 좋음    | 규칙이 복잡할때 좋음         |
